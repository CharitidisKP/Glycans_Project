```python
import pandas as pd

# Load the Excel file
input_file = "decreased.xlsx"
output_file = "decreased_with_delta.xlsx"

# Read the Excel file
df = pd.read_excel(input_file)

# Ensure column names are consistent
df.columns = df.columns.str.strip()

# Get the list of variables (columns from C onward)
variables = df.columns[2:]

# Create a new DataFrame to store the deltas
delta_df = df[df['Time point'] == 'T0'].copy()  # Use T0 rows as the base for the delta

# Calculate the delta for each variable
for variable in variables:
    t0_values = df[df['Time point'] == 'T0'][variable].values
    t12_values = df[df['Time point'] == 'T12'][variable].values
    
    # Delta = T12 - T0
    delta_df[variable] = t12_values - t0_values

# Update the Time Point column to indicate deltas
delta_df['Time point'] = 'Delta'

# Save the delta_df to a new Excel file
delta_df.to_excel(output_file, index=False)

print(f"Delta values have been calculated and saved to '{output_file}'.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# Load the delta Excel file
input_file = "decreased_with_delta.xlsx"
output_file = "correlation_results.xlsx"

# Read the Excel file
df = pd.read_excel(input_file)

# Columns C to AT (variables) and AU to BB (clinical parameters)
variables = df.columns[2:38]  # Adjust based on actual column indices
clinical_parameters = df.columns[39:47]  # Adjust based on actual column indices

# Create a DataFrame to store correlation results
correlation_results = []

# Loop through variables and clinical parameters
for var in variables:
    for clin_param in clinical_parameters:
        # Perform Spearman correlation
        rho, p_value = spearmanr(df[var], df[clin_param], nan_policy='omit')
        
        # Append results to the list
        correlation_results.append({
            "Variable": var,
            "Clinical Parameter": clin_param,
            "Spearman R": rho,
            "P-Value": p_value
        })

# Convert results to a DataFrame
correlation_df = pd.DataFrame(correlation_results)

# Save correlation results to an Excel file
correlation_df.to_excel(output_file, index=False)
print(f"Correlation results saved to '{output_file}'.")

# Create a pivot table for the heatmap
heatmap_data = correlation_df.pivot(index="Variable", columns="Clinical Parameter", values="Spearman R")

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", cbar_kws={'label': 'Spearman R'})
plt.title("Spearman Correlation Heatmap")
plt.xlabel("Clinical Parameters")
plt.ylabel("Variables")
plt.tight_layout()

# Save the heatmap as an image
heatmap_file = "correlation_heatmap.png"
plt.savefig(heatmap_file)
print(f"Heatmap saved to '{heatmap_file}'.")

plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu
import statsmodels.api as sm
import statsmodels.formula.api as smf



# Load your dataset
input_file = "increased_with_delta.xlsx"
df = pd.read_excel(input_file)

# Specify variables and outcomes
variables = df.columns[2:46]  # Adjust based on your dataset
outcomes = ['Clinical', 'Renal Impairment']

# Create a results dictionary to store analysis results
results = []

# Loop through each outcome
for outcome in outcomes:
    for var in variables:
        # Perform logistic regression
        model = sm.Logit(df[outcome], sm.add_constant(df[var], has_constant='add'))
        result = model.fit(disp=False)
        
        # Extract logistic regression results
        odds_ratio = np.exp(result.params[1])  # Odds ratio
        p_value = result.pvalues[1]  # P-value
        ci_lower, ci_upper = np.exp(result.conf_int().iloc[1])  # Confidence interval
        
        # Perform Mann-Whitney U Test
        group_0 = df[df[outcome] == 0][var]
        group_1 = df[df[outcome] == 1][var]
        _, mw_p_value = mannwhitneyu(group_0, group_1, alternative='two-sided', nan_policy='omit')
        
        # Store results
        results.append({
            'Outcome': outcome,
            'Variable': var,
            'Odds Ratio': odds_ratio,
            'CI Lower': ci_lower,
            'CI Upper': ci_upper,
            'Logistic P-Value': p_value,
            'Mann-Whitney P-Value': mw_p_value
        })

# Convert results to a DataFrame and save to Excel
results_df = pd.DataFrame(results)
results_file = "binary_outcome_analysis.xlsx"
results_df.to_excel(results_file, index=False)
print(f"Analysis results saved to '{results_file}'.")
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load the dataset
input_file = "increased_with_delta.xlsx"
df = pd.read_excel(input_file)

# Specify the columns for variables and outcomes
variables = df.columns[2:46]  # Adjust column indices as needed
outcomes = ['Clinical', 'Renal Impairment']

# Prepare lists to store results
clinical_results = []
renal_impairment_results = []

# Loop through each outcome
for outcome in outcomes:
    results = []  # Temporary list for each outcome's results
    for var in variables:
        # Logistic regression model
        model = sm.Logit(df[outcome], sm.add_constant(df[var]))
        result = model.fit(disp=False)
        
        # Extract logistic regression results
        odds_ratio = np.exp(result.params[1])  # Odds ratio
        p_value = result.pvalues[1]  # p-value
        ci_lower, ci_upper = np.exp(result.conf_int().iloc[1])  # Confidence interval
        
        # Append results for plotting
        results.append({
            'Variable': var,
            'Odds Ratio': odds_ratio,
            'CI Lower': ci_lower,
            'CI Upper': ci_upper,
            'P-Value': p_value
        })
    
    # Store results for Clinical or Renal Impairment
    if outcome == 'Clinical':
        clinical_results = results
    else:
        renal_impairment_results = results

# Create a forest plot for Clinical
def plot_forest(results, outcome_name):
    variables = [r['Variable'] for r in results]
    odds_ratios = [r['Odds Ratio'] for r in results]
    ci_lowers = [r['CI Lower'] for r in results]
    ci_uppers = [r['CI Upper'] for r in results]
    
    # Create plot
    plt.figure(figsize=(10, len(results) * 0.5))
    plt.errorbar(odds_ratios, range(len(results)), xerr=[np.array(odds_ratios) - np.array(ci_lowers), np.array(ci_uppers) - np.array(odds_ratios)],
                 fmt='o', color='black', label='Odds Ratio')
    plt.axvline(x=1, color='gray', linestyle='--')  # Reference line for OR = 1
    
    # Add labels
    plt.yticks(range(len(results)), variables)
    plt.xlabel('Odds Ratio (95% CI)')
    plt.title(f"Forest Plot for {outcome_name} Outcome")
    plt.grid(True)
    
    # Show p-values
    for i, result in enumerate(results):
        p_value = result['P-Value']
        plt.text(odds_ratios[i] + 0.1, i, f"p={p_value:.3f}", verticalalignment='center', fontsize=8)
    
    # Show plot
    plt.tight_layout()
    plt.show()

# Plot forest plot for Clinical
plot_forest(clinical_results, 'Clinical')

# Plot forest plot for Renal Impairment
plot_forest(renal_impairment_results, 'Renal Impairment')
```
